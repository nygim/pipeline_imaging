```python
import json

# Replace 'your_file.json' with the path to your JSON file
file_path = "/Users/nayoonkim/pipeline_imaging/file_map.json"

# Read JSON content
with open(file_path, "r") as file:
    data = json.load(file)
```

```python
# Count how many of these elements have "UAB" in the 'input_file' name
uw_count = sum(1 for element in data if "UW" in element["input_file"])
uab_count = sum(1 for element in data if "UAB" in element["input_file"])
ucsd_count = sum(1 for element in data if "UCSD" in element["input_file"])

# Print the count
print(f"Number of files with in the name: {uw_count}")
print(f"Number of files with in the name: {uab_count}")
print(f"Number of files with in the name: {ucsd_count}")
```

```python
# Filter elements that have 'output_files' as an empty list
elements_with_empty_output_files = [
    element
    for element in data
    if "output_files" in element and not element["output_files"]
]

# Count how many of these elements have "UAB" in the 'input_file' name
uab_count = sum(
    1 for element in elements_with_empty_output_files if "UW" in element["input_file"]
)

# Print the count
print(f"Number of files with in the name: {uab_count}")
```
